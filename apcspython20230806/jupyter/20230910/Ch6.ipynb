```python
places = {'A':(10, ''), 'B':(11, ''), 'C':(12, ''), 'D':(13, ''),
          'E':(14, ''), 'F':(15, ''), 'G':(16, ''), 'H':(17, ''),
          'I':(34, ''), 'J':(18, ''), 'K':(19, ''), 'M':(21, ''),
          'N':(22, ''), 'O':(35, ''), 'P':(23, ''), 'Q':(24, ''),
          'T':(27, ''), 'U':(28, ''), 'V':(29, ''), 'W':(32, ''),
          'X':(30, ''), 'Z':(33, ''),
          'L':(20, ''), 'R':(25, ''), 'S':(26, ''), 'Y':(31, '')}

weight = [8,7,6,5,4,3,2,1,1]

def isValidId(pid):
    if(pid[0] not in places):
        return False
    
    if(pid[1]!='1' and pid[1]!='2'):
        return False
    
    checkSum = 0
    checkSum = p2sum(places[pid[0]][0])#
    for i in range(1, 10):
        checkSum += int(pid[i])*weight[i-1]
    if(checkSum%10==0):
        return True
    else:
        return False
# num//10 
#num%10
def p2sum(num):
    return num//10*1+num%10*9

pid = input(':')

if(isValidId(pid)):
    gender = int(pid[1])
    birth = pid[0]
    if(gender == 1):
        print('!')
    else:
        print('!')
    print(':', places[birth][1])
else:
    print(':%s' %pid)
```
